```python
import pandas as pd
```

```python
districts = ['assembly','senate','congressional']
```

```python
def readtracts(d):
    dist = pd.read_csv('blocks-{}districts.csv'.format(d), header=None, skiprows=2)
    dist.columns = ['countyfp','tract','block','legdist_{}'.format(d[0]),'county','population','alloc']
    return dist.drop('alloc', axis=1).set_index(['countyfp','tract','block','county','population'])
```

```python
data = {d: readtracts(d) for d in districts}
```

```python
electeds = pd.read_excel('california-electeds.xlsx', sheetname=None)
```

```python
def readelecteds(sheet):
    df = electeds[sheet][['District','Name','Party']]
    df['Party'] = df['Party'].str.slice(stop=1)
    df['chamber'] = sheet[:-2]
    df['session'] = '20' + sheet[-2:]
    return df
```

```python
districts = pd.concat([readelecteds(s) for s in electeds.keys()])
```
